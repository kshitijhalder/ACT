Plot a per day graph of stock price of Apple in May 2024


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data from the excel file
df = pd.read_excel('/Users/kshitijhalder/Desktop/ACT/Python/June/AppleStocks.xlsx')

# Create a seaborn style
sns.set_theme("notebook", font_scale=1.5, style="darkgrid", palette="deep", color_codes=True, rc=None)
sns.relplot(data=df, x='Date', y='Price', kind='line', aspect=2, height=6, ci=None, markers=True, dashes=False, color='b')
plt.title('Apple Stock Prices')

# If price decreases from the previous day, color the data point red
for i in range(1, len(df)):
    if df['Price'][i] < df['Price'][i-1]:
        plt.plot(df['Date'][i], df['Price'][i], 'ro')

# Give a text message for every time price decreases from previous date
for i in range(1, len(df)):
    if df['Price'][i] < df['Price'][i-1]:
        print(f'Price decreased on {df["Date"][i]}')

# If the price decreases from the previous day, give a notification
for i in range(1, len(df)):
    if df['Price'][i] < df['Price'][i-1]:
        plt.text(df['Date'][i], df['Price'][i], 'Price Decreased', fontsize=12, color='red')

plt.show()
```
