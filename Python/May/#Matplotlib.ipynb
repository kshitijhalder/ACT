Create Dataframe of 15 Students, their marks and attendance and use matplotlib package to create graphs

```python
import pandas as pd
import matplotlib.pyplot as plt

Data = {'Batch 1': ['Dhara', 'Yamini', 'Mahendl', 'Kshitij', 'Het', 'Rishi', 'Abhay', 'Eshan', 'Rajesh', 'Rahul', 'Krishna', 'Kevin', 'Maneet', 'Aditya', 'Akshay'], 'Marks': [90, 80, 70, 60, 50, 40, 30, 20, 10, 100, 90, 80, 70, 60, 50], 'Attendance' : ['100%', '90%', '80%', '70%', '60%', '50%', '40%', '30%', '20%', '10%', '100%', '90%', '80%', '70%', '60%']}
df = pd.DataFrame(Data)
print(df)

# Plot a scatter plot of marks of the students
# Name of the students on x-axis and marks on y-axis
# Only 1st letter of the name should be displayed on x-axis
plt.scatter(df['Batch 1'].str[0], df['Marks'])
#plt.hist(df['Batch 1'],edgecolor='black')
plt.xlabel('Batch 1')
plt.ylabel('Marks')
plt.title('Marks of students')
plt.show()

df['Attendance'] = df['Attendance'].str.replace('%', '')
plt.pie(df['Attendance'] )
#plt.pie(df['Batch 1'].str[0], df['Attendance'])
#plt.xlabel('Batch 1')
#plt.ylabel('Attendance')
plt.title('Attendance of students')
plt.show()

plt.scatter(df['Marks'], df['Attendance'])
plt.xlabel('Marks')
plt.ylabel('Attendance')
plt.title('Marks vs Attendance of students')
plt.show()
```
