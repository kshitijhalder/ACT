Import a data from notepad. Create a dictionary of it therefore a dataframe. Then create a graph

```python
# Import a data from notepad. Create a dictionary of it therefore a dataframe. Then create a graph
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Read the data from the notepad
file = open('/Users/kshitijhalder/Desktop/ACT/Python/May/Notepad.txt', 'r')
data = file.readlines()

# Create an empty dictionary
my_dict = {}

# Process each line of the file
for line in data:
    # Split the line into key and values
    line = line.strip().split()
    key = line[0]
    values = [value.replace(',', '') for value in line[1:]]
    
    # Add key-value pair to the dictionary
    my_dict[key] = values

# Convert marks and attendance to integers and floats respectively
my_dict['Marks'] = [int(marks) for marks in my_dict['Marks']]
my_dict['Attendance'] = [float(attendance.rstrip('%')) for attendance in my_dict['Attendance']]

print(my_dict)

# Create a DataFrame from the dictionary
df = pd.DataFrame(my_dict)
print('\n',df)

# Plot a double bar graph using matplotlib
df.plot(x='Name', y=['Marks', 'Attendance'], kind='bar')

plt.show()

# Plot a double bar graph using seaborn
df_melt = df.melt(id_vars='Name', value_vars=['Marks', 'Attendance'])
sns.barplot(x='Name', y='value', hue='variable', data=df_melt)

plt.show()

# Save the graph to a csv file
df.to_csv('/Users/kshitijhalder/Desktop/ACT/Python/May/Notepad.csv', index=False)

csv = pd.read_csv('/Users/kshitijhalder/Desktop/ACT/Python/May/Notepad.csv')
#print(csv)

# Use of iloc
x = csv.iloc[:, 0:2]
print(x)

# Use of loc
y = csv.loc[:, 'Marks':'Attendance']
print(y)
```

Difference between iloc and loc:

- iloc is used to access the data using the index of the rows and columns
- loc is used to access the data using the labels of the rows and columns
